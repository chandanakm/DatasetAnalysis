```

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

data = pd.read_csv('library electricity.csv')


data1 = data["DateTime_Measured"].str.split(" ", n = 1 ,expand = True)
data["date"]= data1[0]
data["time"]= data1[1]
data.drop(columns =["DateTime_Measured"], inplace = True) 


data2 = data["date"].str.split("-", n = 1 ,expand = True)
data["year"]= data2[0]
data["monthday"]=data2[1]
data.drop(columns =["date"], inplace = True) 




yearw=pd.DataFrame(data['year'].value_counts())
nyearw=yearw.sort_index()
nyearw

data.head()

```

```

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np


plt.bar(nyearw.index,nyearw.year)
plt.xlabel('year', fontsize=15)
plt.ylabel('No of kw', fontsize=15)
plt.title('central library electricity usage')
plt.show()
```

```
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np



data3 = data["monthday"].str.split("-", n = 1 ,expand = True)
data["month"]= data3[0]
data["day"]=data3[1]
data.drop(columns =["monthday"], inplace = True) 
data.head()

```

```
data_2019=data[data.year=="2019"]


#groupedmnth=data_2019.groupby("mnth")
#groupedmnth.groups
mnthw=pd.DataFrame(data_2019['month'].value_counts())
nmnthw=mnthw.sort_index()
nmnthw
```

```
plt.scatter(nmnthw.index,nmnthw.month, 10)
# 10 is going to set the size of the dots in the scatter plot
plt.title('Usage of electricity in library monthwise')
plt.xlabel('months of the year 2019 till july')
plt.ylabel('Demand in kWs')
plt.show()
```
